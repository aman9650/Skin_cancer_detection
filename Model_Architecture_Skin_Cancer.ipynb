## Creating Folder label which contain images label wise


```python

# Path to the folder containing the extracted images
extracted_images_folder = 'extracted_images'

# Iterate over each row in the DataFrame
for index, row in data.iterrows():
    image_id = row['image_id']
    label = row['dx']  # Assuming the label column is named 'label' in the CSV

    # Construct the path to the input image
    input_image_path = os.path.join(extracted_images_folder, f'{image_id}.jpg')

    # Check if the image exists in the extracted images folder
    if os.path.exists(input_image_path):
        # Create a new folder for the label if it doesn't exist already
        output_folder = os.path.join('labels', label)
        os.makedirs(output_folder, exist_ok=True)

        # Copy the image to the folder corresponding to its label
        output_image_path = os.path.join(output_folder, f'{image_id}.jpg')
        shutil.copyfile(input_image_path, output_image_path)
        print(f"Image {image_id} saved with label {label}.")
    else:
        print(f"Image {image_id} not found in extracted images folder. Skipping.")

print("Images saved with labels successfully.")
```

```python

# Path to the labels folder
labels_folder = 'labels'

# Get a list of all subfolders (labels) within the labels folder
labels = [label for label in os.listdir(labels_folder) if os.path.isdir(os.path.join(labels_folder, label))]

# Iterate over each label and count the number of images in the label folder
for label in labels:
    label_folder_path = os.path.join(labels_folder, label)
    num_images = len([name for name in os.listdir(label_folder_path) if os.path.isfile(os.path.join(label_folder_path, name))])
    print(f"Label '{label}' contains {num_images} images.")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array

# Path to the directory containing the images
data_dir = '/content/labels'  # Update with the path to your labels directory

# Create an ImageDataGenerator instance for data augmentation
datagen = ImageDataGenerator(
    rotation_range=20,      # Random rotation within the range of [-20, 20] degrees
    width_shift_range=0.1,  # Random horizontal shift within the range of [-0.1, 0.1] of the total width
    height_shift_range=0.1, # Random vertical shift within the range of [-0.1, 0.1] of the total height
    shear_range=0.2,        # Random shear transformation within the range of [-0.2, 0.2]
    zoom_range=0.2,         # Random zoom within the range of [0.8, 1.2]
    horizontal_flip=True,   # Random horizontal flipping
    vertical_flip=True,     # Random vertical flipping
    fill_mode='nearest'     # Fill mode for filling in newly created pixels
)

# Iterate over each label directory
for label in os.listdir(data_dir):
    if label != 'nv':
        label_dir = os.path.join(data_dir, label)

        # Get list of images in the label directory
        image_files = [os.path.join(label_dir, file) for file in os.listdir(label_dir) if file.endswith('.jpg')]

        # Perform data augmentation for each image and save the augmented images
        for img_path in image_files:
            img = load_img(img_path)  # Load the image
            img = img_to_array(img)   # Convert to numpy array
            img = img.reshape((1,) + img.shape)  # Reshape to (1, height, width, channels)

            # Generate augmented images
            i = 0
            for batch in datagen.flow(img, batch_size=1, save_to_dir=label_dir, save_prefix=label, save_format='jpg'):
                i += 1
                if i >= 5:  # Generate 5 augmented images per original image
                    break  # Exit the loop to move to the next image

print("Data augmentation applied successfully.")
```

```python

# Path to the labels folder
labels_folder = 'labels'

# Get a list of all subfolders (labels) within the labels folder
labels = [label for label in os.listdir(labels_folder) if os.path.isdir(os.path.join(labels_folder, label))]

# Iterate over each label and count the number of images in the label folder
for label in labels:
    label_folder_path = os.path.join(labels_folder, label)
    num_images = len([name for name in os.listdir(label_folder_path) if os.path.isfile(os.path.join(label_folder_path, name))])
    print(f"Label '{label}' contains {num_images} images.")
```

```python
# Define parameters
img_height = 224
img_width = 224
batch_size = 32
num_classes = 7
```

```python
# Create training dataset
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    '/content/labels',
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size
)
```

```python
# Create validation dataset
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    '/content/labels',
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size
)
```

```python
# Normalize pixel values to [0, 1]
train_ds = train_ds.map(lambda x, y: (x / 255.0, y))
val_ds = val_ds.map(lambda x, y: (x / 255.0, y))
```

```python
#-------------------------------model------------------------------
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.layers import Conv2D, MaxPool2D, Dropout, Flatten
from keras.callbacks import ReduceLROnPlateau
from keras.models import Model
from tensorflow.keras.applications import MobileNet
```

```python
# Load MobileNet model
base_model = MobileNet(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))

# Freeze base model
base_model.trainable = False
```

```python
# Add classification head
inputs = tf.keras.Input(shape=(img_height, img_width, 3))
x = base_model(inputs, training=False)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
outputs = tf.keras.layers.Dense(num_classes, activation='softmax')(x)
model = tf.keras.Model(inputs, outputs)
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Define learning rate reduction callback
learning_rate_reduction = ReduceLROnPlateau(monitor='val_accuracy', patience=3, verbose=1, factor=0.5, min_lr=0.00001)
```

```python
# Train the model
history = model.fit(train_ds, epochs=10, validation_data=val_ds, callbacks=[learning_rate_reduction])
```

```python
# Assuming 'model' is your trained model
model.save("my_model.h5")
```
